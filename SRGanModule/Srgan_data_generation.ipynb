```python
import os
import random
from PIL import Image, ImageEnhance, ImageFilter

def create_low_res_variation(input_image_path, output_folder):
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Open the image using PIL
    image = Image.open(input_image_path)

    # List of possible transformations
    possible_transformations = ['night', 'blur', 'noise', 'low_resolution']

    # Randomly choose transformations
    num_transformations = random.randint(1, len(possible_transformations))  # Random number of transformations
    chosen_transformations = random.sample(possible_transformations, num_transformations)

    # Apply chosen transformations
    transformed_image = image.copy()
    for effect in chosen_transformations:
        if effect == 'night':
            # Simulate nighttime effect by decreasing brightness
            transformed_image = ImageEnhance.Brightness(transformed_image).enhance(0.5)
        elif effect == 'blur':
            # Apply Gaussian blur
            transformed_image = transformed_image.filter(ImageFilter.GaussianBlur(radius=2))
        elif effect == 'noise':
            # Add random noise
            noise_factor = random.uniform(0.5, 1.5)
            transformed_image = ImageEnhance.Color(transformed_image).enhance(noise_factor)
        elif effect == 'low_resolution':
            # Resize image to low resolution
            scale_factor = random.randint(2, 8)  # Random scale factor between 2 and 8
            width, height = transformed_image.size
            new_width = int(width / scale_factor)
            new_height = int(height / scale_factor)
            transformed_image = transformed_image.resize((new_width, new_height), Image.LANCZOS)

    # Save the transformed image
    output_image_path = os.path.join(output_folder, f'{os.path.basename(input_image_path)}')
    transformed_image.save(output_image_path)

# Example usage
input_folder_mobile_crane = 'dataset/mobile_crane'
output_folder_low_res_variations_mobile_crane = 'dataset/low_res_variations_mobile_crane'
input_folder_tower_crane = 'dataset/tower_crane'
output_folder_low_res_variations_tower_crane = 'dataset/low_res_variations_tower_crane'

# Process each image in mobile crane folder
files_mobile_crane = os.listdir(input_folder_mobile_crane)
for file in files_mobile_crane:
    input_image_path = os.path.join(input_folder_mobile_crane, file)
    create_low_res_variation(input_image_path, output_folder_low_res_variations_mobile_crane)

# Process each image in tower crane folder
files_tower_crane = os.listdir(input_folder_tower_crane)
for file in files_tower_crane:
    input_image_path = os.path.join(input_folder_tower_crane, file)
    create_low_res_variation(input_image_path, output_folder_low_res_variations_tower_crane)
```
